## Iterative convergence

```python
cell_list = ["0deg", "8deg"]
```

```python
conv_steps_van = np.array([450838, 434142])
conv_steps = conv_steps_van
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 3.5))
for (i, cell) in enumerate(cell_list):
    n = i%3;
    resi = np.loadtxt(path + "residual-{}.txt".format(cell))[0:int(conv_steps[i])]
    axs[n].semilogy(np.arange(0,int(conv_steps[i])), resi[:,0]/resi[0,0], 'C0', label="mass")
    axs[n].semilogy(np.arange(0,int(conv_steps[i])), resi[:,1]/resi[0,1], 'C1', label="x-mtm")
    axs[n].semilogy(np.arange(0,int(conv_steps[i])), resi[:,2]/resi[0,2], 'C2', label="y-mtm")
    axs[n].semilogy(np.arange(0,int(conv_steps[i])), resi[:,3]/resi[0,3], 'C3', label="energy")
    axs[n].grid()
    axs[n].xaxis.set_tick_params(labelsize=14)
    axs[n].yaxis.set_tick_params(labelsize=14)
    axs[n].set_ylim(0.9*10**(-9), 25)
    axs[n].legend(fontsize=14)
    axs[n].set_title("Case: AOA-{}".format(cell),fontsize=15)
plt.savefig("figs/Airfoil_residual.pdf",bbox_inches='tight')
plt.show()
```

## Integral quantity

```python
press_infi = 65855.8
rho_infi = press_infi/(287.0*300.0)
u_infi = np.sqrt(1.4*287.0*300*0.84**2)
```

```python
grid_xy = np.loadtxt(path+"grid.dat",skiprows=5)
grid_xy.shape
```

```python
x_grid = grid_xy[:,0].reshape(14,49)
y_grid = grid_xy[:,1].reshape(14,49)
x_wall = x_grid[0]
y_wall = y_grid[0]
```

```python
x_mid = np.zeros([48])
for i in range(0,48):
    x_mid[i] = 0.5 * (x_wall[i] + x_wall[i+1])
```

```python
my_solu = np.loadtxt(path+"My-solution-0deg.dat",skiprows=5)
my_solu.shape
```

```python
press = my_solu[:,-1].reshape(13,48)
press_1st_layer = press[0]
press_2nd_layer = press[1]
press_wall = (3.0*press_1st_layer - press_2nd_layer)/2.0
```

```python
press_lower = press_wall[8:24]
press_upper = press_wall[24:40]
press_coeff_lower = (press_lower - press_infi)/(0.5*rho_infi*u_infi**2)
press_coeff_upper = (press_upper - press_infi)/(0.5*rho_infi*u_infi**2)
x_lower = x_mid[8:24]/0.1524
x_upper = x_mid[24:40]/0.1524
```

```python
# experiment
x_exp_upper = np.array([0,0.006,0.016,0.027,0.051,0.08,0.106,0.154,0.199,0.255,0.304,0.351,0.399,0.448,0.502,0.551,\
0.6,0.655,0.758,0.804,0.904,0.955,1])
coeff_exp_upper = np.array([1.18,0.35,0.08,-0.08,-0.18,-0.19,-0.28,-0.24,-0.25,-0.3,-0.3,-0.3,-0.34,-0.34,\
                            -0.29,-0.22,-0.16,-0.1,-0.06,0.01,0.09,0.09,0.21])
x_exp_lower = np.array([0.015,0.028,0.052,0.08,0.106,0.154,0.204,0.251,0.352,0.452,0.555,0.655,0.755,0.852,0.955])
coeff_exp_lower = np.array([-0.05,-0.08,-0.04,-0.21,-0.1,-0.2,-0.37,-0.27,-0.3,-0.31,-0.2,-0.15,-0.02,0.08,0.12])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(9.6, 3.0))

axs[0].plot(x_lower,press_coeff_lower,'o',markersize=6,color='red',markeredgewidth=2,markerfacecolor="None",label='Numerical')
axs[0].plot(x_exp_lower,coeff_exp_lower, color='k', linewidth=2, linestyle='--',label='Experimental')
axs[0].set_title("Lower surface")
axs[1].plot(x_upper,press_coeff_upper,'s',markersize=6,color='blue',markeredgewidth=2,markerfacecolor="None",label='Numerical')
axs[1].plot(x_exp_upper,coeff_exp_upper, color='k', linewidth=2, linestyle='--',label='Experimental')
axs[1].set_title("Upper surface")
for i in range(0,2):
    axs[i].grid()
    axs[i].set_ylim(-0.5,1.35)
    axs[i].xaxis.set_tick_params(labelsize=10)
    axs[i].yaxis.set_tick_params(labelsize=10)
    axs[i].legend(fontsize=11)
plt.savefig("figs/Cp-compare.pdf",bbox_inches='tight')
plt.show()
```

### calculate lift and drag coefficient

```python
press_x = np.zeros([48])
press_y = np.zeros([48])
for i in range(0,48):
    press_x[i] = (y_wall[i+1]-y_wall[i])*press_wall[i]
    press_y[i] = -(x_wall[i+1]-x_wall[i])*press_wall[i]
```

```python
Fx = np.sum(press_x[8:24])+np.sum(press_x[24:40])
Fy = np.sum(press_y[8:24])+np.sum(press_y[24:40])
```

```python
Cd = Fx/(0.5*rho_infi*u_infi**2*0.1524)
Cl = Fy/(0.5*rho_infi*u_infi**2*0.1524)
```
