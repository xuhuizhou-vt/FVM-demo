```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams['text.usetex'] = True
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

## 1st order vs. 2nd order

```python
xcell = np.loadtxt("2nd-order-shock/Roe/xcell_40.txt")
xcell_fine = np.loadtxt("2nd-order-shock/Roe/xcell_100.txt")

V_van_1st = np.loadtxt("1st-order-shock/Van-Leer/V_40.txt")
V_van_2nd = np.loadtxt("2nd-order-shock/Van-Leer/V_40.txt")
V_roe_1st = np.loadtxt("1st-order-shock/Roe/V_40.txt")
V_roe_2nd = np.loadtxt("2nd-order-shock/Roe/V_40.txt")

M_van_1st = np.loadtxt("1st-order-shock/Van-Leer/Ma_40.txt")
M_van_2nd = np.loadtxt("2nd-order-shock/Van-Leer/Ma_40.txt")
M_van_2nd_fine = np.loadtxt("2nd-order-shock/Van-Leer/Ma_100.txt")
M_roe_1st = np.loadtxt("1st-order-shock/Roe/Ma_40.txt")
M_roe_2nd = np.loadtxt("2nd-order-shock/Roe/Ma_40.txt")
M_roe_2nd_fine = np.loadtxt("2nd-order-shock/Roe/Ma_100.txt")
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 5))
fig.subplots_adjust(hspace=0.55)
axs[0].plot(xcell, M_van_1st[2:-2], 'C2-', label='1st order (coarse)', linewidth=2)
axs[0].plot(xcell, M_van_2nd[2:-2], 'C1-.',label='2nd order (coarse)', linewidth=2)
axs[0].plot(xcell_fine, M_van_2nd_fine[2:-2], 'C0--',label='2nd order (fine)', linewidth=2)
axs[0].set_title("Van Leer FVS Scheme")

axs[1].plot(xcell, M_roe_1st[2:-2], 'C2-', label='1st order (coarse)', linewidth=2.0)
axs[1].plot(xcell, M_roe_2nd[2:-2], 'C1-.', label='2nd order (coarse)', linewidth=2.0)
axs[1].plot(xcell_fine, M_roe_2nd_fine[2:-2], 'C0--',label='2nd order (fine)', linewidth=2)
axs[1].set_title("Roe's FDS Scheme")

for i in range(0,2):
    axs[i].set_xlabel(r"$x$",fontsize=13)
    axs[i].set_ylabel(r"$M$",fontsize=13)
    axs[i].set_ylim(0,3)
    axs[i].legend(fontsize=10)
    axs[i].grid(linewidth=1, alpha=0.6)

axs[0].tick_params(axis='x', labelsize=13)
axs[0].tick_params(axis='y', labelsize=13)
axs[1].tick_params(axis='x', labelsize=13)
axs[1].tick_params(axis='y', labelsize=13)
plt.savefig("figs/Mach_1st-vs-2nd.pdf",bbox_inches='tight')
plt.show()
```

## different $\kappa$ for 2nd order

### resolution: 100

```python
# van leer
M_van_neg_1p0 = np.loadtxt("2nd-order-shock-diff-kappa/Van-Leer/Ma_100_neg-1p0.txt")
M_van_0p0 = np.loadtxt("2nd-order-shock-diff-kappa/Van-Leer/Ma_100_0p0.txt")
M_van_0p5 = np.loadtxt("2nd-order-shock-diff-kappa/Van-Leer/Ma_100_0p5.txt")
M_van_1p0 = np.loadtxt("2nd-order-shock-diff-kappa/Van-Leer/Ma_100_1p0.txt")
# Roe
M_roe_neg_1p0 = np.loadtxt("2nd-order-shock-diff-kappa/Roe/Ma_100_neg-1p0.txt")
M_roe_0p0 = np.loadtxt("2nd-order-shock-diff-kappa/Roe/Ma_100_0p0.txt")
M_roe_0p5 = np.loadtxt("2nd-order-shock-diff-kappa/Roe/Ma_100_0p5.txt")
M_roe_1p0 = np.loadtxt("2nd-order-shock-diff-kappa/Roe/Ma_100_1p0.txt")
```

```python
# cell
xcell = np.loadtxt("2nd-order-shock-diff-kappa/Roe/xcell_100.txt")
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 5))
fig.subplots_adjust(hspace=0.55)
axs[0].plot(xcell, M_van_neg_1p0[2:-2], 'C2-', label=r'$\kappa=-1.0$', linewidth=2)
axs[0].plot(xcell, M_van_0p0[2:-2], 'C1-.',label=r'$\kappa=0.0$', linewidth=2)
axs[0].plot(xcell, M_van_0p5[2:-2], 'C0--',label=r'$\kappa=0.5$', linewidth=2)
axs[0].plot(xcell, M_van_1p0[2:-2], 'C7--',label=r'$\kappa=1.0$', linewidth=2)
axs[0].set_title("Van Leer FVS Scheme (2nd order)")

axs[1].plot(xcell, M_roe_neg_1p0[2:-2], 'C2-', label=r'$\kappa=-1.0$', linewidth=2)
axs[1].plot(xcell, M_roe_0p0[2:-2], 'C1-.',label=r'$\kappa=0.0$', linewidth=2)
axs[1].plot(xcell, M_roe_0p5[2:-2], 'C0--',label=r'$\kappa=0.5$', linewidth=2)
axs[1].plot(xcell, M_roe_1p0[2:-2], 'C7--',label=r'$\kappa=1.0$', linewidth=2)
axs[1].set_title("Roe's FDS Scheme (2nd order)")

for i in range(0,2):
    axs[i].set_xlabel(r"$x$",fontsize=13)
    axs[i].set_ylabel(r"$M$",fontsize=13)
    axs[i].set_ylim(0,3)
    axs[i].legend(fontsize=11)
    axs[i].grid(linewidth=1, alpha=0.6)

axs[0].tick_params(axis='x', labelsize=12)
axs[0].tick_params(axis='y', labelsize=12)
axs[1].tick_params(axis='x', labelsize=12)
axs[1].tick_params(axis='y', labelsize=12)
plt.savefig("figs/diff-kappa-compare.pdf",bbox_inches='tight')
plt.show()
```

## compare with Jameson's artificial dissipation

```python
M_van = np.loadtxt("2nd-order-shock/Van-Leer/Ma_100.txt")
M_roe = np.loadtxt("2nd-order-shock/Roe/Ma_100.txt")
M_jam = np.loadtxt("../HW2/case2/Ma_100.txt")
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 5))
fig.subplots_adjust(hspace=0.55)
axs[0].plot(xcell, M_van[2:-2], 'C2-', label='Van Leer', linewidth=2)
axs[0].plot(xcell, M_roe[2:-2], 'C1--',label='Roe', linewidth=2)
axs[0].plot(xcell, M_jam[2:-2], 'C0-.',label='w/ damping', linewidth=2)
axs[0].set_title("Different scheme comparison")

axs[1].plot(xcell, M_van[2:-2], 'C2-', label='Van Leer', linewidth=2)
axs[1].plot(xcell, M_roe[2:-2], 'C1--',label='Roe', linewidth=2)
axs[1].plot(xcell, M_jam[2:-2], 'C0-.',label='w/ damping', linewidth=2)
axs[1].set_title("Zoom-in comparison")
axs[1].set_xlim(0.3,0.7)
axs[1].set_xticks([0.3,0.4,0.5,0.6,0.7])

for i in range(0,2):
    axs[i].set_xlabel(r"$x$",fontsize=13)
    axs[i].set_ylabel(r"$M$",fontsize=13)
    axs[i].set_ylim(0,3)
    axs[i].legend(fontsize=12)
    axs[i].grid(linewidth=1, alpha=0.6)
    axs[i].tick_params(axis='x', labelsize=12)
    axs[i].tick_params(axis='y', labelsize=12)

plt.savefig("figs/diff-scheme-comparison.pdf",bbox_inches='tight')
plt.show()
```
